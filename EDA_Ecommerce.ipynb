**Dataset:** ecommerce_sales.csv | 35 rows × 15 columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
%matplotlib inline

df = pd.read_csv('ecommerce_sales.csv')
df['order_date'] = pd.to_datetime(df['order_date'])
df['revenue'] = df['quantity'] * df['unit_price'] * (1 - df['discount_pct']/100)
df['returned_flag'] = (df['returned'] == 'Yes').astype(int)
df['month'] = df['order_date'].dt.month_name()
print('Shape:', df.shape)
```

## 2. Univariate Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

sns.histplot(df['age'], kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('Age Distribution')

sns.histplot(df['revenue'], kde=True, ax=axes[1], color='coral')
axes[1].set_title('Revenue Distribution')

sns.histplot(df['rating'], kde=True, ax=axes[2], color='seagreen')
axes[2].set_title('Rating Distribution')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

df['category'].value_counts().plot(kind='bar', ax=axes[0], color='mediumpurple', edgecolor='black')
axes[0].set_title('Orders by Category')
axes[0].tick_params(axis='x', rotation=30)

df['payment_method'].value_counts().plot(kind='bar', ax=axes[1], color='darkorange', edgecolor='black')
axes[1].set_title('Payment Method Frequency')
axes[1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

## 3. Bivariate Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

df.groupby('category')['revenue'].sum().sort_values().plot(
    kind='barh', ax=axes[0], color='teal', edgecolor='black')
axes[0].set_title('Total Revenue by Category')
axes[0].set_xlabel('Revenue (₹)')

df.groupby('city')['revenue'].sum().sort_values().plot(
    kind='barh', ax=axes[1], color='tomato', edgecolor='black')
axes[1].set_title('Total Revenue by City')
axes[1].set_xlabel('Revenue (₹)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='delivery_days', y='rating', hue='returned', data=df,
                palette={'Yes':'red','No':'green'}, s=80)
plt.title('Delivery Days vs Rating (coloured by Return)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='category', y='revenue', data=df, palette='Set2')
plt.title('Revenue Distribution per Category')
plt.xticks(rotation=30)
plt.show()
```

## 4. Categorical Insights

```python
print('Return Rate by Category:')
print((df.groupby('category')['returned_flag'].mean() * 100).round(1).sort_values(ascending=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[0],
                                  colors=['steelblue','salmon'], startangle=90)
axes[0].set_title('Gender Split')
axes[0].set_ylabel('')

df.groupby('gender')['revenue'].mean().plot(kind='bar', ax=axes[1],
                                             color=['steelblue','salmon'], edgecolor='black')
axes[1].set_title('Avg Revenue by Gender')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

## 5. Correlation Heatmap

```python
plt.figure(figsize=(9, 6))
num_cols = ['age','quantity','unit_price','discount_pct','delivery_days','rating','revenue','returned_flag']
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

## 6. Time Trend

```python
daily = df.groupby('order_date')['revenue'].sum().reset_index()
plt.figure(figsize=(12, 4))
plt.plot(daily['order_date'], daily['revenue'], marker='o', color='royalblue')
plt.title('Daily Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Revenue (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 7. Summary Findings

```python
print('=== KEY FINDINGS ===')
print(f"Total Revenue: ₹{df['revenue'].sum():,.0f}")
print(f"Avg Order Value: ₹{df['revenue'].mean():,.0f}")
print(f"Overall Return Rate: {df['returned_flag'].mean()*100:.1f}%")
print(f"Avg Rating: {df['rating'].mean():.2f}")
print(f"Top Category: {df.groupby('category')['revenue'].sum().idxmax()}")
print(f"Top City: {df.groupby('city')['revenue'].sum().idxmax()}")
print(f"Most Used Payment: {df['payment_method'].value_counts().idxmax()}")
```
